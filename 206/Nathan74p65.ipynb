Une société de voiture propose à ses clients deux contrats:

- Contrat $C_1$ : un forfait de $23$ € et $0,40$ € par kilomètre parcouru;    
- Contrat $C_2$ : $0,60$ € par kilomètre parcouru.

1°) Ecrire un algorithme permettant de calculer et d'afficher le coût des deux contrats suivant le nombre $x$ de kilomètres parcourus.


```python
x=float(input('Saisir le nbre de km : '))
c1=0.4*x+23
c2=0.6*x
print()
print('le coût du contrat C1 pour', x , 'km est :',c1)
print()
print('le coût du contrat C2 pour', x , 'km est :',c2)
```

2°) a) Exprimer le coût de chaque contrat en fonction de x. Et représenter graphiquement les deux fonctions obtenues dans un même repère.

On a $\;f_1(x)=0,4\times x+23\;$ et $\;f_2(x)=0,6\times x\;$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 250, 100)
plt.plot(x, 0.4*x+23, label='f1')
plt.plot(x, 0.6*x, label='f2')

plt.xlabel('distance parcourue (km)')
plt.ylabel('coût (€)')
plt.title('Comparaison des coûts')
plt.legend()
plt.grid()
plt.savefig("test.pdf")
plt.show()
```

2°) b) Graphiquement: Jusqu'à **115 km** le contrat $C_2$ est le plus avantageux.

**Remarque :** on peut aussi utiliser la librairie sympy (calcul symbolique) pour résoudre l'équation $\;f_1(x)=f_2(x)\;$ avec la commande solve.

```python
from sympy import *
x = Symbol('x')
#init_printing(use_latex=True)
solve(Eq(0.4*x+23, 0.6*x), x)
```

2°) c) Ecrire un algorithme dont l'objectif est de donner, suivant le nombre x de kilomètres à parcourir, le nom du contrat le moins cher et le prix à payer.

```python
x=float(input('Saisir le nbre de km : '))
if 0<=x and x<=115:
    P=0.6*x
    contrat='C1'   # sdqsdqdsdsdqsd
else:
    P=0.4*x+23
    contrat='C2'
print("Le contrat le moins cher est : ", contrat)
print("Le montant à payer sera de : ", P, "euros")
```

**Complément :** Avec une boucle "for" pour afficher une liste de résultats

```python
L=[5,10,50,100,115,130,150,200]
for x in L:
    if 0<=x and x<=115:
        P=0.6*x
        contrat='C1'
    else:
        P=0.4*x+23
        contrat='C2'
    print('Pour x = ',x)    
    print("Le contrat le moins cher est : ", contrat)
    print("Le montant à payer sera de : ", P, "euros")
    print('_________________________________________')
```
